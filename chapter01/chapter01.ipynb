#00. 文字列の逆順

```
# 文字列”stressed”の文字を逆に（末尾から先頭に向かって）並べた文字列を得よ．
str = 'stressed'
ans = ""
for i in range(len(str)):
  ans += str[-i - 1]
print(ans) 
```

02. 「パトカー」＋「タクシー」＝「パタトクカシーー」

```
# 「パトカー」＋「タクシー」の文字を先頭から交互に連結して文字列「パタトクカシーー」を得よ．
str1 = 'パトカー'
str2 = 'タクシー'
# リスト内包表記
ans = ''.join(i + j for i, j in zip(str1, str2) ) # ()内で1つのリスト化したものをjoinで連結
                                                  # +：文字列を連結, join()：１つのリストの中身を連結
print(ans)
```

03. 円周率

```
# “Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics.”
#  という文を単語に分解し，各単語の（アルファベットの）文字数を先頭から出現順に並べたリストを作成せよ
import re
str = 'Now I need a drink, alcoholic of course, after the heavy lectures involving quantum mechanics.'
splits = str.split()                       
ans = [len(s.strip(', .')) for s in splits] # リスト内包表記 (len(i)を要素とするリストを作成)
                                            # strip()は文字列の先頭と末尾から任意の文字列を削除
print(ans)
```

04. 元素記号

```
# “Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.”
# という文を単語に分解し，1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字，それ以外の単語は先頭に2文字を取り出し，
# 取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ．
```

```
import re
str = 'Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.'
splits = str.split()  # 文章を空白で分割してリスト(配列)化
one_ch = [1, 5, 6, 7, 8, 9, 15, 16, 19 ]
ans = {} # 辞書の作成
for i, word in enumerate(splits):  # enumerateで配列のインデックスと要素を取り出す
  if i + 1 in one_ch:
    ans[word[:1]] = i + 1    # word[:1]で文字列wordの最初の１文字を取得。それをキーとして、値i+1を代入

  else:
    ans[word[:2]] = i + 1

print(ans)
```

05. n-gram

```
"""
    与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．
    この関数を用い，”I am an NLPer”という文から単語bi-gram，文字bi-gramを得よ．
"""
                                                 # n-gramモデル：文中から連続するn文字を取り出す
def ngram(n, lst):
  return list(zip(*[lst[i:] for i in range(n)])) # [lst[i:] for i in range(n)]で、文中の0～n要素目始まりの文のリストを作成['I am an NLPer',  ' am an NLPer', ...]
                                                 # リストに*をつけて引数に渡すと、展開されてそれぞれの要素が個別の引数として渡される
                                                 #   *['I am an NLPer',  'am an NLPer'] -> ('I am an NLPer',  ' am an NLPer') 
                                                 # list( zip() )で、zipの括弧内の各要素をそれぞれ１つずつ取り出してタプルにまとめたリストを取得
                                                 #  list( zip( ('I am an NLPer',  ' am an NLPer') ) )-> [('I', ' '), (' ', 'a'), ('a', 'm'), ...]
                                                 #                      A                B              (Aの0,Bの0),(Aの1,Bの1), (Aの3,Bの3),...

str = 'I am an NLPer'
words_bi_gram = ngram(2, str.split())
chars_bi_gram = ngram(2, str)

print('単語bi-gram: ',words_bi_gram)
print('文字bi-gram: ', chars_bi_gram)
```

```
def n_gram(target, n):
  return [ target[idx:idx + n] for idx in range(len(target) - n + 1)]  # range(n) は 0～n-1 までのn個

target = 'I am an NLPer'
words = target.split(' ')
print(n_gram(words, 3))
```

6. 集合

```
"""
    “paraparaparadise”と”paragraph”に含まれる文字bi-gramの集合を，それぞれ, XとYとして求め，XとYの和集合，積集合，差集合を求めよ．
      さらに，'se’というbi-gramがXおよびYに含まれるかどうかを調べよ．
"""
str1 = 'paraparaparadise'
str2 = 'paragraph'
X = set(ngram(2, str1))  # set()で、重複する要素が除外されるsetオブジェクトが生成される
Y = set(ngram(2, str2))
union = X | Y           # 和集合
intersection = X & Y    # 積集合
difference = X - Y      # 差集合

print('X:', X)
print('Y:', Y)
print('和集合:', union)
print('積集合:', intersection)
print('差集合:', difference)
print('Xにseが含まれるか:', {('s', 'e')} <= X) # <=で、Xの部分集合か判定。要素を波括弧{}でくくるとset型オブジェクトになる
print('Yにseが含まれるか:', {('s', 'e')} <= Y)
```

#07. テンプレートによる文生成

引数x, y, zを受け取り「x時のyはz」という文字列を返す関数を実装せよ．

さらに，x=12, y=”気温”, z=22.4として，実行結果を確認せよ．


```
def generate_sentence(x, y, z):
  return '{}時の{}は{}'.format(x, y, z)  # 文字列に変換 。f'{x}時の{y}は{z}'　でも可

generate_sentence(12, '気温', 22.4)
```

# 08. 暗号文

> 与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．
>
>英小文字ならば(219 - 文字コード)の文字に置換
>
>その他の文字はそのまま出力
>
>この関数を用い，英語のメッセージを暗号化・復号化せよ．



```
def cipher(str):
  rep = [chr(219 - ord(x)) if x.islower() else x for x in str] # chr()で文字コードから文字を取得。ord()で文字から文字コードを取得。islower()で小文字か判定。
  return ''.join(rep)                                          #「条件式が真のときの式 if 条件式 else 条件式が偽の時に評価される式」
                                                               # 「A if 条件式 else B for x in range(n)」->     for x in range(n)
                                                               #                                                  if 条件式 :
                                                               #                                                  A
                                                               #                                                  else:
                                                               #                                                  B  


message = 'the quick brown fox jumps over the lazy dog'
message = cipher(message)
print('暗号化:', message)
message = cipher(message)
print('複合化:', message)
```

# 09. Typoglycemia


> スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．ただし，長さが４以下の単語は並び替えないこととする．
>
>適当な英語の文（例えば”I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .”）を与え，その実行結果を確認せよ



```
import random

def shuffle(words):
  result = []
  for word in words.split():
    if len(word) > 4:
      word = word[:1] + ''.join(random.sample(word[1:-1], len(word[1:-1]))) + word[-1:]  # word[:1]は最初の文字。word[-1:]は末尾の文字
    result.append(word)                         # append()でリストに要素を追加           # word[1:-1]は最初と末尾の間の範囲の文字
  return ' '.join(result)                       # random.sample(リスト, 取得したい要素数)は元のリストを変更せず、ランダムに並び替えた新しいリストを作成
  

words = "I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."
ans = shuffle(words)
print(ans)
```
